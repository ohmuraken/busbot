```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect("timetable.db")
con.row_factory = sqlite3.Row
timetable = pd.read_csv("timetable.csv")
```

```python
timetable
```

```python
sql = """
    create table Timetable(
        id integer primary key,
        lines integer,
        hour integer,
        minute integer,
        wednesday integer
    );
"""
con.execute(sql)
con.commit()
```

```python
time_data = list(zip(
        timetable["lines"],
        timetable["hour"],
        timetable["minute"], 
        timetable["wednesday"]
    ))
# time_data
```

```python
for l, h, m, w in time_data:
    sql = "insert into Timetable(lines,hour,minute,wednesday) values({}, {}, {},{})".format(l, h, m,w)
    con.execute(sql)
    con.commit()
```

```python
con.close()
```

```python
hour = 10
c = con.execute(u"select * from Timetable where hour > {}".format(hour))
for row in c:
    print("{}".format(row["hour"]))
```
